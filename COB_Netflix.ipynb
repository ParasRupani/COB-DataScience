```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset.csv')

# Missing values
df.fillna('Not Given', inplace=True)

# Updating date format
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['date_added'] = df['date_added'].dt.strftime('%d/%m/%Y')

# Update duration in numerical format
def format_duration(duration):
    try:
        return int(''.join(filter(str.isdigit, duration)))
    except:
        return None

df['duration'] = df['duration'].apply(format_duration)
```

```python
# Outliers in duration

Q1 = df['duration'].quantile(0.25) #(25th percentile)
Q3 = df['duration'].quantile(0.75) #(75th percentile)
IQR = Q3 - Q1 #InterQuartile
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_duration = df[(df['duration'] < lower_bound) | (df['duration'] > upper_bound)]

print(f'Number of outliers: {len(outliers_duration)}')

# Removing Outliers from the dataset
df_cleaned = df[~((df['duration'] < lower_bound) |
                           (df['duration'] > upper_bound))]
```

```python
# Outliers in release_year

Q1 = df['release_year'].quantile(0.25)
Q3 = df['release_year'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_year = df[(df['release_year'] < lower_bound) | (df['release_year'] > upper_bound)]

print(f'Number of outliers: {len(outliers_year)}')

# Removing Outliers from the dataset
df_cleaned_year = df_cleaned[~((df_cleaned['release_year']
                                    < lower_bound)
                                    | (df_cleaned['release_year']
                                    > upper_bound))]
```
